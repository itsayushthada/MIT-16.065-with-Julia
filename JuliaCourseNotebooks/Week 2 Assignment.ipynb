# Student-T Distribution's Probability Density Function

```julia
function pdf(x, nu::Int64)
    t1 = gamma((nu+1)/2)
    t2 = (1 + ((x^2)/nu))^(-(nu+1)/2)
    t3 = sqrt(nu*pi) * gamma(nu/2)
    
    return t1 * t2 / t3
end
```

$$ pdf(x) = 
\frac
{
    \Gamma\left(\frac{\nu+1}{2}\right) . 
        \left(1+\frac{x^2}{\nu}\right) ^{-\frac{\left(\nu+1\right)}{2}}
}
{
    \sqrt{\nu\pi} . \Gamma\left(\frac{\nu}{2}\right)
}
$$ 
<br>
Here,<br>
$\nu$ denotes the degrees of freedom.<br>
$\Gamma$ represents gamma function.

The t-Distribution, also known as Studentâ€™s t-Distribution is the probability distribution that estimates the population parameters when the sample size is small and the population standard deviation is unknown.

It resembles the normal distribution and as the sample size increases the t-distribution looks more normally distributed with the values of means and standard deviation of 0 and 1 respectively.

```julia
X = Array{Int64, 1}(undef, 21)
Y1 = Array{Float64, 1}(undef, 21)
Y2 = Array{Float64, 1}(undef, 21)
Y3 = Array{Float64, 1}(undef, 21)
Y4 = Array{Float64, 1}(undef, 21)
Y5 = Array{Float64, 1}(undef, 21)
Y6 = Array{Float64, 1}(undef, 21)

for x = -10:10
    X[x+11] = x
    Y1[x+11] = pdf(x,1)
    Y2[x+11] = pdf(x,2)
    Y3[x+11] = pdf(x,5)
    Y4[x+11] = pdf(x,10)
    Y5[x+11] = pdf(x,30)
    Y6[x+11] = pdf(x,15000)
end
```

```julia
Plots.plot( X, 
            [Y1, Y2, Y3, Y4, Y5, Y6],
            label = ["dof-1", "dof-2", "dof-5", "dof-10", "dof-30", "dof-15000"],
            title = "T Distribution with different Degrees of Freedom",
            xlabel = "Random Variable",
            ylabel = "Probability",
            linewidth = 2,
            size = (900, 600))
```
